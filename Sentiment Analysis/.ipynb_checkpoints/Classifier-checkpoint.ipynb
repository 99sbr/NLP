```python
def clean_tweet(tweet):
    tweet = tweet.lower()
    tweet = tweet.replace("#","")
    tweet = re.sub(r'http\S+', '',tweet)
    tweet = "".join(i for i in tweet if ord(i)<128)
    tweet = tweet.replace("[^a-zA-Z0-9]", " ")
    tweet = [wordninja.split(word) for word in tweet.split()]
    tweet = " ".join([item for items in tweet for item in items])
    return tweet
```

```python
# WORD2VEC
from gensim.models import FastText
W2V_SIZE = 300
W2V_WINDOW = 5
W2V_EPOCH = 32
W2V_MIN_COUNT = 1

f2t_model = FastText(size=W2V_SIZE,
                     window=W2V_WINDOW,
                     min_count=W2V_MIN_COUNT,
                     workers=4, iter=1000)
```

```python
%%time
documents = [_text.split() for _text in train_df.clean_tweet]
```

```python
%%time
f2t_model.build_vocab(documents)
```

```python
%%time
words = f2t_model.wv.vocab.keys()
vocab_size = len(words)
print("Vocab size", vocab_size)
f2t_model.train(documents, total_examples=len(documents), epochs=W2V_EPOCH)
```

```python
embed_dim = 300 # how big is each word vector
max_features = vocab_size # how many unique words to use (i.e num rows in embedding vector)
max_seq_len = max_seq_len # max number of words in a question to use
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Activation, Dense, Dropout, Embedding, Flatten, Conv1D, MaxPooling1D, LSTM,Input
from keras import utils
from keras.callbacks import ReduceLROnPlateau, EarlyStopping , CSVLogger
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, TensorBoard
from keras.layers import LSTM, Dense, Embedding, Bidirectional,GlobalMaxPool1D, Lambda , GRU
from keras.layers import TimeDistributed
from keras.layers import *
from keras.optimizers import Adam
```

```python
tokenizer = Tokenizer(num_words=max_features)
tokenizer.fit_on_texts(X_train)
word_index = tokenizer.word_index
word_seq_train = tokenizer.texts_to_sequences(X_train)
word_seq_test = tokenizer.texts_to_sequences(X_test)
word_index = tokenizer.word_index
print("dictionary size: ", len(word_index))

#pad sequences
word_seq_train = pad_sequences(word_seq_train, maxlen=max_seq_len)
word_seq_test = pad_sequences(word_seq_test, maxlen=max_seq_len)
```

```python
y_train = y_train.values.reshape(-1,1)
print("y_train",y_train.shape)

y_test = y_test.values.reshape(-1,1)
print("y_test",y_test.shape)
```

```python

print('preparing embedding matrix...')
words_not_found = []
nb_words = min(max_features, len(word_index)+1)
embedding_matrix = np.zeros((nb_words, embed_dim))
for word, i in word_index.items():
    if i >= nb_words:
        continue
    embedding_vector = f2t_model.wv.get_vector(word)
    if (embedding_vector is not None) and len(embedding_vector) > 0:
        # words not found in embedding index will be all-zeros.
        embedding_matrix[i] = embedding_vector
    else:
        words_not_found.append(word)
print('number of null word embeddings: %d' % np.sum(np.sum(embedding_matrix, axis=1) == 0))
```

```python
embedding_layer = Embedding(nb_words, 300, weights=[embedding_matrix], input_length=max_seq_len, trainable=False)
```

```python
import keras.backend as K
```

```python
# build network
from keras.utils.generic_utils import get_custom_objects
class Swish(Activation):
    
    def __init__(self, activation, **kwargs):
        super(Swish, self).__init__(activation, **kwargs)
        self.__name__ = 'swish'
        
def swish(x):
    return K.sigmoid(x) * x
```

```python
model = Sequential()
get_custom_objects().update({'swish': Swish(swish)})
model.add(embedding_layer)
model.add(LSTM(256))
model.add(Dense(64))
model.add(Activation('swish'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
%%time
history = model.fit(word_seq_train, y_train,
                    batch_size=32,
                    epochs=5,
                    validation_data=(word_seq_test, y_test),
                    verbose=1,shuffle=True)
```
